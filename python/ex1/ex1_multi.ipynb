```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('ggplot')
plt.xkcd()
```

```python
data = np.loadtxt('ex1data2.txt', delimiter=',', usecols=(0, 1, 2), unpack=True)

# Initialize feature and label data
X = np.transpose(np.array(data[:-1]))
y = np.transpose(np.array(data[-1:]))
m = y.size
```

```python
# Normalize features
def featureNormalize(X):
    mu = np.mean(X, axis=0)
    X_norm = X - mu
    
    sigma = np.std(X_norm, axis=0)
    X_norm = X_norm / sigma
    
    return X_norm, mu, sigma

X, mu, sigma = featureNormalize(X)

# Add ones to first column of X
X = np.insert(X, 0, 1, axis=1)
```

```python
# Gradient Descent

# Vectorized solutions generalize for any sized X, theta as long as
# number of X columns equals number of theta rows

def computeCost(X, y, theta):
    return float(1. / (2*m) * (X @ theta - y).T @ (X @ theta - y))

def gradientDescent(X, y, theta, alpha, num_iters):
    m = y.size
    J_history = []
    theta_history = []
    
    for i in range(num_iters):
        delta = (alpha/m) * (X.T @ (X @ theta - y))
        theta = theta - delta
        J_history.append(computeCost(X, y, theta))
        theta_history.append(theta)
    
    return theta, J_history, theta_history

alpha = 0.01
num_iters = 400
theta_init = np.zeros((X.shape[1], 1))

# Run gradient descent on differing values of alpha
theta_one, J_one, _ = gradientDescent(X, y, theta_init, alpha, num_iters);
theta_two, J_two, _ = gradientDescent(X, y, theta_init, 0.03, num_iters);
theta_three, J_three, _ = gradientDescent(X, y, theta_init, 0.005, num_iters);
```

```python
# Comparing Gradient Descent with different values of alpha

plt.figure(figsize=(10, 6))

plt.plot(J_one, '-b', linewidth=2, label='alpha = 0.01')
plt.plot(J_two, '-r', linewidth=2, label='alpha = 0.03')
plt.plot(J_three, '-g', linewidth=2, label='alpha = 0.005')

plt.legend()
plt.xlabel('Iterations');
plt.ylabel('Cost J');

plt.show()
```

```python
# Estimating the price of a 1650 sq-foot, 3 bedroom house

# Using alpha = 0.01
theta = theta_one

price = np.insert((np.array([1650, 3]) - mu) / sigma, 0, 1, axis=0) @ theta

print("Predicted price of a 1650 sq-ft, 3 bedroom house: $%.2f" % float(price))
```

```python
# Normal equations

def normalEqn(X, y):
    return np.linalg.pinv(X.T @ X) @ X.T @ y # theta

# Reload variables
data = np.loadtxt('ex1data2.txt', delimiter=',', usecols=(0, 1, 2), unpack=True)

# Initialize feature and label data
X = np.transpose(np.array(data[:-1]))
y = np.transpose(np.array(data[-1:]))
m = y.size

X = np.insert(X, 0, 1, axis=1)

theta = normalEqn(X, y)

print('Theta computed from normal equation: \n{}\n'.format(theta))

price = np.array([1, 1650, 3]) @ theta

print('Predicted price of a 1650 sq-ft house with 3 bedrooms using normal equations: $%.2f' % price)
```
