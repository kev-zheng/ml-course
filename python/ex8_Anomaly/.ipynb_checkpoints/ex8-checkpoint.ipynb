```python
"""
This notebook introduces anomaly detection by using
the gaussian and multivariate gaussian distribution
and F1 score to detect outliers
"""

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import seaborn as sns

import scipy.io as sio
import math

sns.set_context('notebook')
#sns.set_style('white')
plt.style.use('fivethirtyeight') # blue
#plt.style.use('ggplot') # red
plt.xkcd() # because why not?
np.set_printoptions(precision=3)
np.seterr(divide='ignore', invalid='ignore') # Ignore divide by zero
```

```python
def visualizeFit(X, mu, sigma2):
    """Plots contours that correspond with the multivariate gaussian distribution"""
    interval = np.arange(0, 35, 0.5)
    X1, X2 = np.meshgrid(interval, interval)

    Z = multivariateGaussian(np.array([X1.flatten(), X2.flatten()]).T, mu, sigma2).reshape(X1.shape)
    
    plt.scatter(X[:, 0], X[:, 1], edgecolors='k',linewidths=1)
    plt.xlabel('Latency (ms)')
    plt.ylabel('Throughput (mb/s)')
    
    # Plot only if the values are finite
    if np.any(np.isinf(Z)) == False:
        plt.contour(X1, X2, Z, 10.**(np.arange(-20,0,3)).T, cmap=cm.jet, linewidths=1)
        plt.axis([0, 30, 0, 30])
```

```python
def estimateGaussian(X):
    """Estimates the parameters of a gaussian distribution on X"""
    m, n = X.shape
    mu = np.mean(X, 0)
    sigma2 = 1/m * np.sum((X - mu)**2, 0)
    return mu, sigma2

def multivariateGaussian(X, mu, sigma2):
    """Estimates the parameters of a multivariate gaussian distribution on X"""
    k = len(mu)
    
    # If sigma2 is a vector, turn it into a covariance matrix
    if sigma2.ndim == 1:
        sigma2 = np.diag(sigma2)
        
    X = X - mu
    p = (2 * math.pi) ** (-k/2) * np.linalg.det(sigma2)**(-0.5) * \
        np.exp(-0.5 * np.sum((X @ np.linalg.pinv(sigma2)) * X, 1))
    return p

def selectThreshold(yval, pval):
    """Determines the best threshold epsilon based on F1 score"""
    step = (pval.max() - pval.min()) / 1000
    yval = yval.flatten()
    
    best_F1 = 0
    best_epsilon = 0
    
    for epsilon in np.arange(pval.min(), pval.max(), step):
        pred = pval < epsilon
        
        true_pos = np.sum(pred & yval)
        false_pos = np.sum(pred * (yval==0))
        false_neg = np.sum((pred==0) & yval)
        
        prec = true_pos / (true_pos + false_pos)
        rec = true_pos / (true_pos + false_neg)
        
        F1 = (2*prec*rec)/(prec+rec)
        
        if F1 > best_F1:
            best_epsilon = epsilon
            best_F1 = F1
    
    return best_epsilon, best_F1 
```

```python
data = sio.loadmat('ex8data1.mat')
data.keys()
```

```python
X = data['X']
Xval = data['Xval']
yval = data['yval']
```

```python
plt.scatter(X[:, 0], X[:, 1], edgecolors='k',linewidths=1)
plt.xlabel('Latency (ms)')
plt.ylabel('Throughput (mb/s)')
plt.axis([0, 30, 0, 30]);
```

```python
# Visualizing Gaussian Fit

# Assume a gaussian distribution to the dataset
mu, sigma2 = estimateGaussian(X)

p = multivariateGaussian(X, mu, sigma2)

plt.figure(figsize=(10, 8))
visualizeFit(X, mu, sigma2)
```

```python
# Determining a good epsilon threshold by testing against cross validation set
pval = multivariateGaussian(Xval, mu, sigma2);
epsilon, F1 = selectThreshold(yval, pval);
outliers = np.where(p < epsilon);
```

```python
plt.figure(figsize=(10, 8))

visualizeFit(X, mu, sigma2)

plt.scatter(X[outliers, 0], X[outliers, 1], s=350, facecolors='none', edgecolors='orange', linewidth=1)

#plt.plot(X[outliers, 0], X[outliers, 1], 'o', markersize=20, markerfacecolor="None", markeredgecolor='None')

plt.show()
```

```python
data = sio.loadmat('ex8data2.mat')
data.keys()
```

```python
X = data['X']
Xval = data['Xval']
yval = data['yval']
```

```python
# Estimating outliers on a multi-featured dataset

mu, sigma2 = estimateGaussian(X)

p = multivariateGaussian(X, mu, sigma2);
pval = multivariateGaussian(Xval, mu, sigma2);

epsilon, F1 = selectThreshold(yval, pval)

outliers = np.sum(p < epsilon)

print('Best epsilon on cross validation: {0:0.3g} with a F1 of {1:0.3g}'.format(epsilon, F1))
print('# of outliers found: {}'.format(outliers))
```
